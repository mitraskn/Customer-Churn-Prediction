```python
import pandas as pd
x=pd.read_csv("customer_data.csv")
x['TotalCharges']=pd.to_numeric(x['TotalCharges'],errors='coerce')
# this converts column to numeric and columns like (" ") cant be converted into so use 'error=coerce'  which converts that into 'NaN'
#similarly there 'error=raise' and 'error=ignore' ,can be used based on the situation
x['TotalCharges'].dtype # checks whether all converted into that particular data type
```

```python
#EDA -exploratory data analysis using seaborn
#countplot()-count categories
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=pd.read_csv("customer_data.csv")
sns.countplot(x="Churn",data=x)
plt.show()
```

```python
#histplot()- distribution
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=pd.read_csv("customer_data.csv")
sns.histplot(x["MonthlyCharges"])
plt.show()
```

```python
#boxplot() - to compare groups
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=pd.read_csv("customer_data.csv")
sns.boxplot(x="Churn",y="MonthlyCharges",data=x)
plt.show()
# thid image shows =>Customers who churn('yes') generally have higher monthly charges , than who doesnt churn ('No')
```

```python
# heatmap() - shows the relationship between columns
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=pd.read_csv("customer_data.csv")
corr=x.corr(numeric_only=True) # create a correlation matrix with numeric colums alone
sns.heatmap(corr,annot=True) # draw heatmap and 'annot=True' - to show correlation values inside box 
plt.show()
```
